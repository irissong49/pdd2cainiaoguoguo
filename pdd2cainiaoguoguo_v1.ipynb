```python
#############################################################################
### PDD罪大恶极！ 店家怨声载道！ 谁放着菜鸟裹裹的便宜邮费不用用你的六块钱运费啊！###
#############################################################################

#给想使用菜鸟裹裹商家版发货的拼多多商家使用

#-通过网页版拼多多复制地址，拼出菜鸟裹裹批量发货的xlxs，上传进行批量发货
#-安全绿色无公害的本地剪贴板读取，成功读取之后阿米娅语音提示防止误触。


#点击cell运行，去网页版点复制地址，复制“发货”俩字或者等一段时间会自动结束，在OUTPUT_PATH生成xlsx，然后自己上传到菜鸟果果那儿发货
#务必填写最前面的configuration里的各种path

#之后有空再做成exe吧



from IPython.display import Audio
import pandas as pd
import pyperclip


########CONFIGURE HERE！！！！###############
XLSX_PATH=r"C:\Users\username\Downloads\cainiaoguoguomodel.xlsx" # model excel path from cainiaoguoguo
OUTPUT_PATH=r"C:\Users\username\Downloads\output.xlsx" #output path

SE1 = r'C:\Users\username\Downloads\明日方舟语音\阿米娅\选中干员1.wav'# sound effect for finished copying
SE_END=r"C:\Users\username\Downloads\明日方舟语音\阿米娅\非3星结束行动.wav" #sound effecct for start making xlsx
########CONFIGURE HERE！！！！###############

audio1 = Audio(SE1, autoplay=True)
audio_end = Audio(SE_END, autoplay=True)

TIMEOUT=3600000

def read_excel(file_path):
    try:
        df = pd.read_excel(file_path)
        return df
    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
    except Exception as e:
        print(f"An error occurred while reading the Excel file: {e}")

file_path = XLSX_PATH
df = read_excel(file_path)




timer=0
address_list=[]

prev_clipboard=""
pyperclip.copy(prev_clipboard)
clipboard_text = pyperclip.paste()




while True:
    if timer > TIMEOUT:
        display(audio_end)
        break
    
    timer+=1
    
    clipboard_text = pyperclip.paste().replace('\r',"")
    if clipboard_text==prev_clipboard:
        #not copying new content, ignore
        pass
    elif "发货" in clipboard_text:
        display(audio_end)
        break
    else:
        #copying new content, save to file
        address_list.append(clipboard_text)
        prev_clipboard=clipboard_text
        print("copied from clip board:\n\t"+clipboard_text)
        display(audio1)

        

print("----------")
print("Copying finished, please check if there is anything wrong")
print(address_list)
print("----converting inputs to excel----")


address_list_clean=[]
for add in address_list:
    address_list_clean.append(add.split('\n'))


def write_excel_row(df,i,ReceiverName,ReceiverPhone,ReceiverAddress,
                    PackageType="日用品",PackageContent="",Comment=""):
    new_row_df = pd.DataFrame([])
    df = pd.concat([df, new_row_df], ignore_index=True)
    df.loc[i] = [i, ReceiverName, ReceiverPhone, ReceiverAddress, PackageType, PackageContent, Comment]


    return df


for i,address in enumerate(address_list_clean):
    [ReceiverName,ReceiverPhone,ReceiverAddress]=address
    df=write_excel_row(df,i+1,ReceiverName,ReceiverPhone,ReceiverAddress)


def save_to_excel(df, output_file):
    try:
        df.to_excel(output_file, index=False)
        print(f"DataFrame saved to {output_file}")
    except Exception as e:
        print(f"An error occurred while saving DataFrame to Excel: {e}")

save_to_excel(df,OUTPUT_PATH)
```
